```python
def suma(a,b):
    res = a+b
    return print("El resultado es",res)
def resta(a,b):
    res = a-b
    return print("El resultado es",res)
def div(a,b):
    res = a/b
    return print("El resultado es",res)
def mult(a,b):
    res = a*b
    return print("El resultado es",res)
def exp(a,b):
    res = a ** b
    return print("El resultado es",res)
def raiz(a):
    res = a ** 0.5
    return print("El resultado es",res)
    
print("------------------Bienvenido a Calculadora------------------\n Suma(1) | Resta(2) | Division(3) | Multiplicacion(4) | Exponencial(5) | Raiz(6) ")
check = int(input("Que problema quiere resolver?"))
a = float(input("Numero 1: "))
b = float(input("Numero 2: "))

if check == 1:
    suma(a,b)
elif check == 2:
    resta(a,b)
elif check == 3:
    div(a,b)
elif check == 4:
    mult(a,b)
elif check == 5:
    exp(a,b)
elif check == 6:
    raiz(a)
```
